```python
import numpy as np
```

```python
A = np.array([1,2,3])
B = np.array([4,3,0])
```

```python
print("A+B = ",np.add(A,B))
print("A-B = ",np.subtract(A,B))
print(np.multiply(A,B))
print(A*B)
```

```python
C = np.array([[1,2,3],[1,1,1]])
D = np.array([[4,3,9,9],[1,1,2,2],[4,2,1,1]])
print(np.dot(C,D))
print(np.matmul(C,D))
```
